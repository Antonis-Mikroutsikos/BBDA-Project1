```python
import pandas as pd

df = pd.read_csv("v6_Latest_Data_Science_Salaries.csv")

df['Categorized Job Title']=' '
for i in range(len(df['Job Title'])):
    j=df['Job Title'][i]
    if j =='Data Engineer':
        df['Categorized Job Title'][i]='Data Engineer'
    elif j =='Data Scientist':
        df['Categorized Job Title'][i]='Data Scientist'
    elif j =='Data Analyst':
        df['Categorized Job Title'][i]='Data Analyst'        
    elif j =='Machine Learning Engineer':
        df['Categorized Job Title'][i]='ML Engineer'  
    elif j =='ML Engineer':
        df['Categorized Job Title'][i]='ML Engineer'
    elif j =='Data Science Manager':
        df['Categorized Job Title'][i]='Data Scientist'  
    elif j =='Applied Scientist':
        df['Categorized Job Title'][i]='Data Scientist'
    elif j =='Machine Learning Scientist':
        df['Categorized Job Title'][i]='ML Engineer'
    elif j =='Research Scientist':
        df['Categorized Job Title'][i]='Data Scientist'  
    elif j =='Data Science Consultant':
        df['Categorized Job Title'][i]='Data Scientist'  
    elif j =='AI Engineer':
        df['Categorized Job Title'][i]='ML Engineer'
    else:
        df['Categorized Job Title'][i]='Other'
```

```python
df=df[~(df['Categorized Job Title']=='Other')].reset_index(drop=True)
```

```python
def country_to_continent(country, continents):
    for countries,continent in continents.items():
        if country in countries:
            return continent
df['Employee Residence per Continent'] = df['Employee Residence'].apply(lambda x: country_to_continent(x, continents))
df['Company Residence per Continent'] = df['Company Location'].apply(lambda x: country_to_continent(x, continents))
```
